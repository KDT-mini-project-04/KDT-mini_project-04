```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
file1 = '시도_합계출산율__모의_연령별_출산율_20240205210029.xlsx'
birth = pd.read_excel(file1, index_col='시도별',)
birth=birth.drop(labels=['시도별'])
birth
```

```python
birth.iloc[1]
```

```python
year = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
plt.title('[2013 - 2022 수도권 출산율]')
#plt.legend('서울특별시', '경기도', '인천광역시')
plt.plot(year, birth.loc['서울특별시'], marker='o', label = '서울특별시')
plt.plot(year, birth.loc['경기도'], marker='o',label = '경기도')
plt.plot(year, birth.loc['인천광역시'], marker='o', label = '인천광역시')
plt.plot(year, birth.loc['세종특별자치시'], marker='o', label = '세종특별자치시')
plt.legend()
plt.show()
```

```python
year = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
plt.title('[2013 - 2022 지방 출산율(도)]')
plt.plot(year, birth.loc['강원도'], marker='o', label = '강원도') 
plt.plot(year, birth.loc['충청북도'], marker='o', label = '충청북도')
plt.plot(year, birth.loc['충청남도'], marker='o', label = '충청남도')
plt.plot(year, birth.loc['전라북도'], marker='o', label = '전라북도')
plt.plot(year, birth.loc['전라남도'], marker='o', label = '전라남도')
plt.plot(year, birth.loc['경상북도'], marker='o', label = '경상북도')
plt.plot(year, birth.loc['경상남도'], marker='o', label = '경상남도')
plt.plot(year, birth.loc['제주특별자치도'], marker='o', label = '제주도')
plt.legend()
```

```python
year = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
plt.title('[2013 -2022 지방 출생율(광역시)]')
plt.plot(year, birth.loc['부산광역시'], marker='o', label = '부산광역시') 
plt.plot(year, birth.loc['대구광역시'], marker='o', label = '대구광역시')
plt.plot(year, birth.loc['광주광역시'], marker='o', label = '광주광역시')
plt.plot(year, birth.loc['대전광역시'], marker='o', label = '대전광역시')
plt.plot(year, birth.loc['울산광역시'], marker='o', label = '울산광역시')
plt.legend()
```
