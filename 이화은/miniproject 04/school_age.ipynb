```python
school_age2.columns = [f'{i}' for i in range(2013, 2023)]
```

```python
school_total = school_age2.xs('총합', level=1)
```

```python
school_total
```

```python
seoul = school_total.iloc[1]
pusan = school_total.iloc[2]
daegu = school_total.iloc[3]
inchen = school_total.iloc[4]
gwangju = school_total.iloc[5]
daejeon = school_total.iloc[6]
ulsan = school_total.iloc[7]
sejeong = school_total.iloc[8]
gg = school_total.iloc[9]
gw = school_total.iloc[10]
cb = school_total.iloc[11]
cn = school_total.iloc[12]
jb = school_total.iloc[13]
jn = school_total.iloc[14]
gb = school_total.iloc[15]
gn = school_total.iloc[16]
jj = school_total.iloc[17]
```

```python
year = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
plt.figure(figsize=(7,7))
plt.plot(year, seoul, marker='o', label = '서울특별시')
plt.plot(year, inchen, marker='o', label = '인천광역시')
plt.plot(year, sejeong, marker='o', label = '세종특별자치시')
plt.plot(year, gg, marker='o', label='경기도')
plt.legend()
plt.title('[2013 - 2022 수도권 학령 인구 추이]')
```

```python
plt.figure(figsize=(7, 7))
plt.plot(year, pusan, marker='o', label = '부산광역시')
plt.plot(year, daegu, marker='o', label = '대구광역시')
plt.plot(year, gwangju, marker='o', label = '광주광역시')
plt.plot(year, daejeon, marker='o', label='대전광역시')
plt.plot(year, ulsan, marker='o', label='울산광역시')
plt.legend()
plt.title('[2013 - 2022 지방(광역시) 학령 인구 추이]')
```

```python
plt.plot(year, gw, marker='o', label='강원도')
plt.plot(year, cb, marker='o', label='충청북도')
plt.plot(year, cn, marker='o', label='충청남도')
plt.plot(year, jb, marker='o', label='전라북도')
plt.plot(year, jn, marker='o', label='전라남도')
plt.plot(year, gb, marker='o', label='경상북도')
plt.plot(year, gn, marker='o', label='경상남도')
plt.plot(year, jj, marker='o', label='제주도')
plt.legend()
plt.title('[2013 - 2022 지방(도) 학령 인구 추이]')
```
